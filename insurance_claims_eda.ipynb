```python


# Function to categorize age
def categorize_age(age):
  if age < 20:
    return '<20'
  elif 21 <= age <= 30:
    return '21-30'
  elif 31 <= age <= 40:
    return '31-40'
  elif 41 <= age <= 50:
    return '41-50'
  elif 51 <= age <= 60:
    return '51-60'
  else:
    return '>60'

# Apply the function to create a new 'age_category' column
df['age_category'] = df['age'].apply(categorize_age)

# Display the value counts for the new category column
print(df['age_category'].value_counts())

# Optional: Display the first few rows with the new column
print(df[['age', 'age_category']].head())
```
